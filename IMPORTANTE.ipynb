```python
n = np.size(Moscu)
y = Moscu
x = fechas
```

```python
a = (n*np.sum(x*y)-np.sum(x)*np.sum(y))/(n*np.sum(x**2) -  np.sum(x)**2)
```

```python
b = np.mean(y)- a*np.mean(x)
```

```python
print(f'La pendiente es: {a}\nLa interseccion es: {b}')
```

```python
y=a*x+b
```

```python
y
```

```python
np.linalg.norm(Moscu - y)
```

```python
plt.plot(fechas,Moscu)
plt.plot(x,y)
plt.plot(25,a*25+b,'og')
plt.plot(75,a*75+b,'or')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar una secuencia de valores de x de -1 a 1
x1 = np.arange(-1, 1, 0.0001)

# Implementación de la fórmula
y1 = np.sqrt(1 - x**2)
y2 = -np.sqrt(1 - x**2)

# Graficar el gráfico con las dos partes del círculo
plt.plot(x1, y1, 'r')
plt.plot(x1, y2, 'r' )

# Agregar el título del gráfico y las etiquetas de los ejes x e y
plt.title("Círculo")
plt.xlabel("Eje x")
plt.ylabel("Eje y")

# Mostrar el gráfico
plt.show()
```
